# YOLO vs. CocoSsd

In this notebook, we will perform object detection using both yolo and cocossd, and make a comparison between them at the end.

```python
from ml5_ipynb import ml5_detector
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_path1 = 'pic/people_ani3.jpg'
img1 = cv2.imread(img_path1)
img_rgba1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGBA)
width = 640
height = 420
resize_size = (width, height)
img_rgba1 = cv2.resize(img_rgba1, resize_size)
x1, y1, channel1 = img_rgba1.shape
print(x1,y1,channel1)
plt.imshow(img_rgba1)
plt.figure()


img_path2 = 'pic/people_ani1.jpg'
img2 = cv2.imread(img_path2)
img_rgba2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGBA)
img_rgba2 = cv2.resize(img_rgba2, resize_size)
x2, y2, channel2 = img_rgba2.shape
print(x2,y2,channel2)
plt.imshow(img_rgba2)
```

### YOLO

```python
yolo = ml5_detector.ObjectDetector('yolo')
```

### CocoSsd

```python
coco = ml5_detector.ObjectDetector('cocossd')
```

### Comparison 

For the image example, yolo only detects the person on the left side while cocossd detects both person and cow (actually it is rhinoceros). In theory, cocossd performs better yolo in multi-object detections, which can be seen in the following results.

```python
def compare(img_array, width, height):
    yolo.image_detect(img_array,width=width,height=height)
    coco.image_detect(img_array,width=width,height=height)
    print("yolo")
    yolo.draw_bounding_box(img_array)
    print("coco")
    coco.draw_bounding_box(img_array)
```

```python
compare(img_rgba1, width = width, height = height)
```

```python
compare(img_rgba2, width = width, height = height)
```
